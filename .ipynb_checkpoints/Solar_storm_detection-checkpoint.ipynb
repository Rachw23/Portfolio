## Script to determine solar storm level as defined by NOAA

Calculates and plots the total integrated proton flux above 10 MeV for a given month from GOES satellite data

Setting up the pandas, numpy and matplotlib routines.

```python
# Import pandas, numpy and matplotlib packages
import warnings
warnings.simplefilter('ignore', FutureWarning)
from pandas import *
from numpy import  *
%matplotlib inline
```

Reading in the variables for the month, number of days in the month, year and number of the GOES satellite for the data used here. The variables are obtained from the user in input boxes. 

Recommendation for this example: year=2015, month=06, days=30, satnumber=13

```python
# Input the year of the data used
YYYY = input('Which year? (Enter 4 digit number)')

# Input the month of the data used
MM = input('Which month? (Enter in MM format)')

# Input the number of days in the month used
DD = input('How many days are in that month?') # i.e. end-date

# Input the number of the GOES satellite number
NN = input('What is the number of the GOES satellite?')
```

The above strings will be used to create new strings below to hold the full URLs of the two csv files, i.e. one for the EPEAD data and one for the HEPAD data.


The columns containing date & time data in both dataframes are then converted into datetime64 format.

```python
# Constructing the strings to hold the URLs of the EPEAD and HEPAD data files
# for the GOES satellite number (NN), month (MM), year (YYYY) and day (DD) strings previously input

# EPEAD URL string
EPEAD = 'https://www.ncei.noaa.gov/data/goes-space-environment-monitor/access/avg/'+YYYY+'/'+MM+'/goes'+NN+'/csv/g'+NN+'_epead_p17ew_5m_'+YYYY+MM+'01_'+YYYY+MM+DD+'.csv'
# Calculating the header row value and storing in variable nEhead
# (row number 672 + number of days input for that specific month)
nEhead = 672 + int(DD)
# Read in the data from CSV to dataframe dfE
dfE = read_csv(EPEAD, header=nEhead)

# HEPAD URL string
HEPAD = 'https://www.ncei.noaa.gov/data/goes-space-environment-monitor/access/avg/'+YYYY+'/'+MM+'/goes'+NN+'/csv/g'+NN+'_hepad_ap_5m_'+YYYY+MM+'01_'+YYYY+MM+DD+'.csv'
# Calculating the header row value and storing in variable nHhead 
# (Row number 254 + number of days input for that specific month)
nHhead = 254 + int(DD)
# Read in the data from CSV to dataframe dfH
dfH = read_csv(HEPAD, header=nHhead)

# Convert each date/time column labelled 'time_tag' into datetime64 format
# Convert EPEAD dataframe's 'time_tag' column to datetime64 format
dfE['time_tag'] = to_datetime(dfE['time_tag'])
# Convert HEPAD dataframe's 'time_tag' column to datetime64 format
dfH['time_tag'] = to_datetime(dfH['time_tag'])
```

```python
# Keep only rows where all quality flags are zero for each dataframe
# For EPEAD, sum all quality flags for each of the east and west facing detectors,
# keeping rows where sum is equal to zero only

# Sum of quality flags for east-facing detector
dfE['TOT_QUALE'] = dfE['P2E_QUAL_FLAG'] + dfE['P3E_QUAL_FLAG'] + dfE['P4E_QUAL_FLAG'] + dfE['P5E_QUAL_FLAG'] + dfE['P6E_QUAL_FLAG'] + dfE['P7E_QUAL_FLAG']

# Sum of quality flags for west-facing detector
dfE['TOT_QUALW'] = dfE['P2W_QUAL_FLAG'] + dfE['P3W_QUAL_FLAG'] + dfE['P4W_QUAL_FLAG'] + dfE['P5W_QUAL_FLAG'] + dfE['P6W_QUAL_FLAG'] + dfE['P7W_QUAL_FLAG']

# Sum of both east and west detectors quality flags
dfE['TOT_QUAL'] = dfE['TOT_QUALE']+dfE['TOT_QUALW']

# Creating new dataframe of rows where the sum of quality flags for both detectors is zero
dfE = dfE[dfE.TOT_QUAL == 0]

# For HEPAD, sum all quality flags
dfH['TOT_QUAL'] = dfH['P8_QUAL_FLAG'] + dfH['P9_QUAL_FLAG'] + dfH['P10_QUAL_FLAG'] + dfH['P11_QUAL_FLAG']

# Creating new dataframe of rows where the sum of quality flags is zero
dfH = dfH[dfH.TOT_QUAL == 0]
```

Taking averages of the East & West corrected fluxes in the P2 - P7 proton channels, only the required columns of times and the proton fluxes are carried forward.

```python
# Calculate the average for the east and west detector corrected fluxes in each proton channel

# Average corrected flux for proton channel P2
dfE['P2_FLUX']=(dfE['P2E_COR_FLUX']+dfE['P2W_COR_FLUX'])/2

# Average corrected flux for proton channel P3
dfE['P3_FLUX']=(dfE['P3E_COR_FLUX']+dfE['P3W_COR_FLUX'])/2

# Average corrected flux for proton channel P4
dfE['P4_FLUX']=(dfE['P4E_COR_FLUX']+dfE['P4W_COR_FLUX'])/2

# Average corrected flux for proton channel P5
dfE['P5_FLUX']=(dfE['P5E_COR_FLUX']+dfE['P5W_COR_FLUX'])/2

# Average corrected flux for proton channel P6
dfE['P6_FLUX']=(dfE['P6E_COR_FLUX']+dfE['P6W_COR_FLUX'])/2

# Average corrected flux for proton channel P7
dfE['P7_FLUX']=(dfE['P7E_COR_FLUX']+dfE['P7W_COR_FLUX'])/2

# Storing only EPEAD average corrected flux across east and west detectors and time data in dataframe
headings = ['time_tag', 'P2_FLUX', 'P3_FLUX', 'P4_FLUX', 'P5_FLUX','P6_FLUX', 'P7_FLUX']
dfE = dfE[headings]

# Storing only HEPAD corrected fluxes for each proton channel and time data in dataframe
headings = ['time_tag', 'P8_FLUX', 'P9_FLUX', 'P10_FLUX', 'P11_FLUX']
dfH = dfH[headings]
```

The EPEAD and HEPAD dataframes are joined on 'time_tag' column with an inner join, and only rows with entries are kept in.

```python
# Using merge command to inner join EPEAD and HEPAD dataframes on the 'time_tag' column,
# only for rows with data in both dataframes
dfEH = merge(dfE, dfH, on='time_tag', how='inner')
```

The effective energy of each proton channel is the geometric mean of the channel boundary energies.

```python
# Store in variables the effective energies of the proton channels (P2-P11) from boundary energies as the geometric mean 
E2 = (4.2*8.7)**0.5
E3 = 11.2
E4 = 24.5
E5 = 55.8
E6 = 130
E7 = 315
E8 = 372
E9 = 463
E10 = 598
E11 = 1000.0
```

The piece-wise power-law slopes and normalisations between each pair of energy values are calculated as:

```python
# Calculate the piece-wise power law slopes and normalisation between energy values 2 & 3
dfEH['alpha_23'] = (log10(dfEH['P3_FLUX'])-log10(dfEH['P2_FLUX'])) / (log10(E3)-log10(E2))
dfEH['A_23'] = 10**(log10(dfEH['P2_FLUX'])-(dfEH['alpha_23']*log10(E2)))

# Calculate the piece-wise power law slopes and normalisation between energy values 3 & 4
dfEH['alpha_34'] = (log10(dfEH['P4_FLUX'])-log10(dfEH['P3_FLUX'])) / (log10(E4)-log10(E3))
dfEH['A_34'] = 10**(log10(dfEH['P3_FLUX'])-(dfEH['alpha_34']*log10(E3)))

# Calculate the piece-wise power law slopes and normalisation between energy values 4 & 5
dfEH['alpha_45'] = (log10(dfEH['P5_FLUX'])-log10(dfEH['P4_FLUX'])) / (log10(E5)-log10(E4))
dfEH['A_45'] = 10**(log10(dfEH['P4_FLUX'])-(dfEH['alpha_45']*log10(E4)))

# Calculate the piece-wise power law slopes and normalisation between energy values 5 & 6
dfEH['alpha_56'] = (log10(dfEH['P6_FLUX'])-log10(dfEH['P5_FLUX'])) / (log10(E6)-log10(E5))
dfEH['A_56'] = 10**(log10(dfEH['P5_FLUX'])-(dfEH['alpha_56']*log10(E5)))

# Calculate the piece-wise power law slopes and normalisation between energy values 6 & 7
dfEH['alpha_67'] = (log10(dfEH['P7_FLUX'])-log10(dfEH['P6_FLUX'])) / (log10(E7)-log10(E6))
dfEH['A_67'] = 10**(log10(dfEH['P6_FLUX'])-(dfEH['alpha_67']*log10(E6)))

# Calculate the piece-wise power law slopes and normalisation between energy values 7 & 8
dfEH['alpha_78'] = (log10(dfEH['P8_FLUX'])-log10(dfEH['P7_FLUX'])) / (log10(E8)-log10(E7))
dfEH['A_78'] = 10**(log10(dfEH['P7_FLUX'])-(dfEH['alpha_78']*log10(E7)))

# Calculate the piece-wise power law slopes and normalisation between energy values 8 & 9
dfEH['alpha_89'] = (log10(dfEH['P9_FLUX'])-log10(dfEH['P8_FLUX'])) / (log10(E9)-log10(E8))
dfEH['A_89'] = 10**(log10(dfEH['P8_FLUX'])-(dfEH['alpha_89']*log10(E8)))

# Calculate the piece-wise power law slopes and normalisation between energy values 9 & 10
dfEH['alpha_910'] = (log10(dfEH['P10_FLUX'])-log10(dfEH['P9_FLUX'])) / (log10(E10)-log10(E9))
dfEH['A_910'] = 10**(log10(dfEH['P9_FLUX'])-(dfEH['alpha_910']*log10(E9)))

# Calculate the piece-wise power law slopes and normalisation between energy values 10 & 11
dfEH['alpha_1011'] = (log10(dfEH['P11_FLUX'])-log10(dfEH['P10_FLUX'])) / (log10(E11)-log10(E10))
dfEH['A_1011'] = 10**(log10(dfEH['P10_FLUX'])-(dfEH['alpha_1011']*log10(E10)))
```

The integrated proton flux across each of the channels from 10 MeV to 1000 MeV are:

```python
# Calculate the integrated proton flux from 10 MeV to 1000 MeV across channels 2 & 3
dfEH['totflux_23'] = (dfEH['A_23']/(dfEH['alpha_23']+1))*(E3**(dfEH['alpha_23']+1)-E2**(dfEH['alpha_23']+1))

# Calculate the integrated proton flux from 10 MeV to 1000 MeV across channels 3 & 4
dfEH['totflux_34'] = (dfEH['A_34']/(dfEH['alpha_34']+1))*(E4**(dfEH['alpha_34']+1)-E3**(dfEH['alpha_34']+1))

# Calculate the integrated proton flux from 10 MeV to 1000 MeV across channels 4 & 5
dfEH['totflux_45'] = (dfEH['A_45']/(dfEH['alpha_45']+1))*(E5**(dfEH['alpha_45']+1)-E4**(dfEH['alpha_45']+1))

# Calculate the integrated proton flux from 10 MeV to 1000 MeV across channels 5 & 6
dfEH['totflux_56'] = (dfEH['A_56']/(dfEH['alpha_56']+1))*(E6**(dfEH['alpha_56']+1)-E5**(dfEH['alpha_56']+1))

# Calculate the integrated proton flux from 10 MeV to 1000 MeV across channels 6 & 7
dfEH['totflux_67'] = (dfEH['A_67']/(dfEH['alpha_67']+1))*(E7**(dfEH['alpha_67']+1)-E6**(dfEH['alpha_67']+1))

# Calculate the integrated proton flux from 10 MeV to 1000 MeV across channels 7 & 8
dfEH['totflux_78'] = (dfEH['A_78']/(dfEH['alpha_78']+1))*(E8**(dfEH['alpha_78']+1)-E7**(dfEH['alpha_78']+1))

# Calculate the integrated proton flux from 10 MeV to 1000 MeV across channels 8 & 9
dfEH['totflux_89'] = (dfEH['A_89']/(dfEH['alpha_89']+1))*(E9**(dfEH['alpha_89']+1)-E8**(dfEH['alpha_89']+1))

# Calculate the integrated proton flux from 10 MeV to 1000 MeV across channels 9 & 10
dfEH['totflux_910'] = (dfEH['A_910']/(dfEH['alpha_910']+1))*(E10**(dfEH['alpha_910']+1)-E9**(dfEH['alpha_910']+1))

# Calculate the integrated proton flux from 10 MeV to 1000 MeV across channels 10 & 11
dfEH['totflux_1011'] = (dfEH['A_1011']/(dfEH['alpha_1011']+1))*(E11**(dfEH['alpha_1011']+1)-E10**(dfEH['alpha_1011']+1))
```

The total integrated proton flux above 10 MeV is obtained by summing the individual channel integrated fluxes, and from this the solar storm level can be determined.

```python
# Sum the integrated fluxes to calculate the total integrated flux above 10 MeV
dfEH['totflux'] = dfEH['totflux_23'] + dfEH['totflux_34'] + dfEH['totflux_45'] + dfEH['totflux_56'] + dfEH['totflux_67'] + dfEH['totflux_78'] + dfEH['totflux_89'] + dfEH['totflux_910'] + dfEH['totflux_1011']

# Determine the solar storm level by taking the logarithm of the total integrated flux above 10 MeV to the base 10
# Round this value down to the next intger with floor method and display
dfEH['storm'] = (log10(dfEH['totflux'])).apply(floor)
dfEH['storm']
```

The maxiumum proton flux for the month, when it happened and what solar storm level it was:

```python
# Determine maximum proton flux where row has total integrated flux over 10 MeV maxiumum value
maxrow=dfEH['totflux'].idxmax()
maxflux=dfEH['totflux'].iloc[maxrow]

# Determine when storm happened - what time corresponds to the maximum value of total integrated flux over 10 MeV
maxtime = dfEH['time_tag'].iloc[maxrow]

# Determine storm level - the value of the logarithm to the base 10 of the total integrated flux above 10 MeV (rounded down)
maxstorm = dfEH['storm'].iloc[maxrow]

# Display all with graph below
```

A graph of the integrated proton flux above 10 MeV versus time is plotted below, using a log scale for the flux axis and changing the index to the 'time_tag' column. 

```python
# Change dataframe index to column for 'time_tag'
dfEH.set_index(['time_tag'], inplace=True)

# Plot graph of integrated proton flux over 10 MeV using log scale flux axis
dfEH.plot(y='totflux', grid=True, logy=True, figsize=(20,5), title='Integrated proton Flux (GOES 13) June 2015', xlabel='Date & Time', ylabel='Total integrated proton flux above 10 MeV', legend=False)

# Display maximum proton flux value
print ('The maximum flux is ', maxflux, '. ')
# Display what time it occured at
print ('The time at which the maximum flux occured was ', maxtime, '. ')
# Display storm level
print ('The storm level is ', maxstorm, '. ')
```
